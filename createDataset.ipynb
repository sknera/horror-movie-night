```python
import pandas as pd
df = pd.read_csv ('horrory - Oceny.csv')
```

```python
df.drop(df.index[:1], inplace=True)
df.drop(df.index[112:], inplace=True)
df.rename(columns = {'Unnamed: 0':'Title'}, inplace = True)
df.set_index('Title', inplace=True)
df
```

```python
#Kto obejrzał WIĘCEJ niż 15 filmów?
people=[]
for i,j in df.count()[1:-2].items():
    if j>15:
        print(i,j)
        people.append(i)
people    
```

```python
df=df[people]
df
```

```python
df.to_csv('dataset.csv', index=True)
```
