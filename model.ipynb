```python
import pandas as pd
df = pd.read_csv('dataset.csv')
df.set_index('Title', inplace=True)
df.dropna(how='all', inplace=True)
```

```python
category = pd.read_csv('category.csv')
category.set_index('Title', inplace=True)
category.astype(int)
```

```python
for person in df:
    new_x=[]
    for x in df[person]:
        if isinstance(x, str):
            if ',' in x:
                x='.'.join(x.split(","))
        new_x.append(x)
    df[person]=new_x
df=df.astype(float)
df
```
